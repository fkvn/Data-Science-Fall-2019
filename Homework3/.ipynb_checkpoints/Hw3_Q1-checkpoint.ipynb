### a.1 - importing needed packages and libraries

```python
import numpy as np
import pandas as pd
```

### a.2 - Read data file "Hearts_s.csv" and assign it to DataFrame

```python
df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")
```

## b - Checking the dataset by printing every 10 lines:

```python
df[0::10]
```

## c - Let’s ignore these categorical features for now, only keep the numerical features and build your feature matrix and label vector.

### c.1 - Creating the Feature Matrix for the df dataset:

```python
# list of feature names that only contain numerical:
feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR', 'Oldpeak']

# use the above list to select the features from the original DataFrame
X = df[feature_cols]  

# print the first 5 rows
X.head()
```

```python
# checking the size of Feature Matix X:

print(X.shape)
```

### c.2 - Select a Series of labels (the last column) from the DataFrame

```python
y = df['AHD']
```

```python
# checking the label vector by printing every 10 values
y[::10]
```

## d - Split the dataset into testing and training sets with the following parameters: test_size=0.25, random_state=6.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, 
                                                    random_state=6)
```

```python
# print the size of the traning set:
print(X_train.shape)
print(X_test.shape)
```

```python
# print the size of the testing set:
print(X_test.shape)
print(y_test.shape)
```

## e - 

### Use KNN (with k=3), Decision Tree (with random_state=5), and Logistic Regression Classifiers to predict Heart Disease based on the training/testing datasets that you built in part (d). 

### Then check, compare, and report the accuracy of these 3 classifiers. Which one is the best? Which one is the worst?

### e.0 - Setting up

```python
# Evaluating the accuracy 
from sklearn.metrics import accuracy_score
```

```python
# Making a list of accuracy for KNN, Decision Tree, and Logistic Regressin
list_of_accuracies = {}
```

### e.1 - Use KNN with k = 3

```python
# Step1: Import KNeighborsClassifier "Class"
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Step2: Instantiate a KNN object with k = 3
k = 3
my_knn_for_hw2 = KNeighborsClassifier(n_neighbors=k) # name of the object is arbitrary!
```

```python
# Step3: Training ONLY the training set
my_knn_for_hw2.fit(X_train, y_train)
```

```python
# Step4: Testing on the testing set
y_predict = my_knn_for_hw2.predict(X_test)
```

```python
# Step5: Evaluating the accuracy 
accuracy =  accuracy_score(y_test, y_predict)

list_of_accuracies['KNN'] = accuracy

print("KNN: ", list_of_accuracies['KNN'])
```

### e.2 - Decision Tree (with random_state=5)

```python
# Step1: Import DecisionTreeClassifier "Class"
from sklearn.tree import DecisionTreeClassifier
```

```python
# Step2: Instantiate a DecisionTree object
my_decisiontree = DecisionTreeClassifier(random_state=5)
```

```python
# print the size of the testing set:
print(X_train.shape)
print(X_test.shape)
```

```python
# Step3: Training ONLY the training set
my_decisiontree.fit(X_train, y_train)
```

```python
# Step4: Testing on the testing set
y_predict_dt = my_decisiontree.predict(X_test)
```

```python
# Step5: Evaluating the accuracy 
accuracy =  accuracy_score(y_test, y_predict_dt)

list_of_accuracies['Decision Tree'] = accuracy

print("Decision Tree: ", list_of_accuracies['Decision Tree'])
```

### e.3 - Logistic Regression Classifiers 

```python
# Step1: Import LogisticRegression "Class"
from sklearn.linear_model import LogisticRegression
```

```python
# Step2: Instantiate a LogisticRegression object
my_logreg = LogisticRegression(solver='liblinear')
```

```python
# Step3: Training ONLY the training set
my_logreg.fit(X_train, y_train)
```

```python
# Step4: Testing on the testing set
y_predict_lr = my_logreg.predict(X_test)
```

```python
# Step5: Evaluating the accuracy 
accuracy =  accuracy_score(y_test, y_predict_lr)

list_of_accuracies['Logistic Regression'] = accuracy

print("Logistic Regression: ", list_of_accuracies['Logistic Regression'])
```

### e.4 - Then check, compare, and report the accuracy of these 3 classifiers. Which one is the best? Which one is the worst?

```python
best_accuracy_lable = max(list_of_accuracies, key=list_of_accuracies.get)
best_accuracy_value = list_of_accuracies[best_accuracy_lable]

worst_accuracy_lable = min(list_of_accuracies, key=list_of_accuracies.get)
worst_accuracy_value = list_of_accuracies[worst_accuracy_lable]

print("The accuracy of 3 classifiers: ", list_of_accuracies)
print("\nThe best accuracy: ", best_accuracy_lable, ": ", best_accuracy_value)
print("\nThe worst accuracy: ", worst_accuracy_lable, ": ", worst_accuracy_value)
```

## g - Repeat parts (d) and (e) with the new dataset that you built in part (f). How does the prediction accuracy change for each method?

### g.1 - Setting up 

```python
# list of all new feature columns
feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR', 'Oldpeak',
                'Gender_f', 'Gender_m', 
                'ChestPain_asymptomatic', 'ChestPain_nonanginal', 'ChestPain_nontypical', 'ChestPain_typical',
                'Thal_fixed', 'Thal_normal', 'Thal_reversable']

# use the above list to select the features from the original DataFrame
X = newDf[feature_cols]  

# Select a Series of labels (the last column) from the DataFrame
y = newDf['AHD']
```

```python
# List of accuracy based on a newDf
list_of_accuracies_newDF = {}
```

```python
# Split the dataset - test_size = 0.25, random_state = 6
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, 
                                                    random_state=6)
```

### g.2 - KNN

```python
# instantiate object
k = 3
my_knn_for_hw2 = KNeighborsClassifier(n_neighbors=k) 

# traning 
my_knn_for_hw2.fit(X_train, y_train)

# testing
y_predict = my_knn_for_hw2.predict(X_test)

# evaculating the accuracy
accuracy =  accuracy_score(y_test, y_predict)

list_of_accuracies_newDF['KNN'] = accuracy

print("KNN: ", list_of_accuracies_newDF['KNN'])

```

### g.3 - Decision Tree

```python
# instantiate object
my_decisiontree = DecisionTreeClassifier(random_state=5)

# traning 
my_decisiontree.fit(X_train, y_train)

# testing
y_predict_dt = my_decisiontree.predict(X_test)

# evaculating the accuracy
accuracy =  accuracy_score(y_test, y_predict_dt)

list_of_accuracies_newDF['Decision Tree'] = accuracy

print("Decision Tree: ", list_of_accuracies_newDF['Decision Tree'])
```

### g.4 - Logistic Regression

```python
# instantiate object
my_logreg = LogisticRegression(solver='liblinear')

# traning 
my_logreg.fit(X_train, y_train)

# testing
y_predict_lr = my_logreg.predict(X_test)

# evaculating the accuracy
accuracy =  accuracy_score(y_test, y_predict_lr)

list_of_accuracies_newDF['Logistic Regression'] = accuracy

print("Logistic Regression: ", list_of_accuracies_newDF['Logistic Regression'])
```

### g.5 - How does the prediction accuracy change for each method

```python
for k, v in list_of_accuracies.items():
    print("\n", k, ": ")
    print("      Original DF: ", list_of_accuracies[k])
    print("      After OneHotCoding: ", list_of_accuracies_newDF[k])
    
```

#### As we can see, the accuracies of both Decision Tree and Logistic Regession are increasing, while KNN's doesn't change

## h -

### Now, repeat part (e) with the new dataset that you built in part (f), but this time using Cross-Validation. 

### Thus, rather than splitting the dataset into testing and training, use 10-fold Cross-Validation (as we learned in Lab4) to evaluate the classification methods and report the final prediction accuracy.

### h.1 - Setting up:

```python
#  importing the method
from sklearn.model_selection import cross_val_score
```

```python
# list of accuracy for Cross-Validation
list_of_accuracies_CV = {}
```

### h.2 - KNN

```python
# instantiate object
k = 3
my_knn_for_hw2 = KNeighborsClassifier(n_neighbors=k) 

# function cross_val_score performs Cross Validation with k = 10:
accuracy_list = cross_val_score(my_knn_for_hw2, X, y, cv=10, scoring='accuracy')

# use average of accuracy values as final result
accuracy_cv = accuracy_list.mean()

list_of_accuracies_CV['KNN'] = accuracy_cv

print("KNN: ", list_of_accuracies_CV['KNN'])
```

### h.3 - Decision Tree

```python
# instantiate object
my_decisiontree = DecisionTreeClassifier(random_state = 5)

# function cross_val_score performs Cross Validation with k = 10:
accuracy_list = cross_val_score(my_decisiontree, X, y, cv=10, scoring='accuracy')

# use average of accuracy values as final result
accuracy_cv = accuracy_list.mean()

list_of_accuracies_CV['Decision Tree'] = accuracy_cv

print("Decision Tree: ", list_of_accuracies_CV['Decision Tree'])
```

### h.4 - Logistic Regression

```python
# instantiate object
my_logreg = LogisticRegression(solver='liblinear')

# function cross_val_score performs Cross Validation with k = 10:
accuracy_list = cross_val_score(my_logreg, X, y, cv=10, scoring='accuracy')

# use average of accuracy values as final result
accuracy_cv = accuracy_list.mean()

list_of_accuracies_CV['Logistic Regression'] = accuracy_cv

print("Logistic Regression: ", list_of_accuracies_CV['Logistic Regression'])
```

### h.5 - Final report

```python
print("K-fold Cross Validation with K = 10: ")
for k, v in list_of_accuracies_CV.items():
    print("\n", k, ": ", list_of_accuracies_CV[k])
```
